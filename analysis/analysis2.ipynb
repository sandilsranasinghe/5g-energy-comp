```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
bsinfo = pd.read_csv("BSinfo.csv")
cldata = pd.read_csv("CLdata.csv")
ecdata = pd.read_csv("ECdata.csv")
prdata = pd.read_csv("to_predict.csv")
```

```python
cldata['Time'] = pd.to_datetime(cldata['Time'])
```

```python
ecdata['Time'] = pd.to_datetime(ecdata['Time'])
```

```python
prdata 
```

```python
prdata = prdata.drop(["w"], axis = 1)
```

```python
combined_dataset = pd.concat([ecdata, prdata], axis=0, ignore_index=True)
```

```python
combined_dataset = combined_dataset.sort_values(by=["BS", "Time"])
```

```python
# window_size = 5
# combined_dataset = combined_dataset.fillna(combined_dataset.rolling(window=2 * window_size + 1, min_periods=1).mean())
```

```python
combined_dataset.fillna(method='ffill', inplace=True)
```

```python
combined_dataset.head(10)
```

```python
# combined_dataset
```

```python
inner_join_result = pd.merge(combined_dataset, prdata, on=["Time","BS"], how='inner')
```

```python
inner_join_result
```

```python
inner_join_result.drop("Energy_y", axis = 1, inplace = True)
```

```python
inner_join_result.rename(columns = {"Energy_x": "Energy"}, inplace = True)
```

```python
inner_join_result
```

```python
inner_join_result['ID'] = inner_join_result['Time'].astype(str) + '_' + inner_join_result['BS']

# Select only the "ID" and "Energy" columns
inner_join_result = inner_join_result[["ID", "Energy"]]
```

```python
inner_join_result
```

```python
inner_join_result.to_csv("power_consumption_prediction.csv", index=False) 
```
